```python
def simple_plot(df_plot):

    plt.figure(figsize= (20,10))
    
#     y = df_plot[df_plot.columns[1]]

    x_p1 = df_plot[(df_plot['time1'] == 1507374600) ][(df_plot.columns[0])]
    y_p1 = df_plot[(df_plot['time1'] == 1507374600) ][(df_plot.columns[1])]
    plt.scatter(x_p1, y_p1, label ='part1 flat' )
    
    x_p2 = df_plot[(df_plot['time1'] == 1512774600)][(df_plot.columns[0])]
    y_p2 = df_plot[(df_plot['time1'] == 1512774600)][(df_plot.columns[1])]
    plt.scatter(x_p2, y_p2, label ='part2 increase', marker = '<'  )
    
    x_p3 = df_plot[(df_plot['time1'] == 1515174300)][(df_plot.columns[0])]
    y_p3 = df_plot[(df_plot['time1'] == 1515174300)][(df_plot.columns[1])]
    plt.scatter(x_p3, y_p3, label ='part3 decrease', marker =  'x')
    
    x_p4 = df_plot[(df_plot['time1'] == 1547873100)][(df_plot.columns[0])]
    y_p4 = df_plot[(df_plot['time1'] == 1547873100)][(df_plot.columns[1])]
    plt.scatter(x_p4, y_p4, label ='part4 flat' , marker = 'o') 
    
    
    plt.xlabel(df_plot.columns[0])
    plt.ylabel(df_plot.columns[1])
    plt.legend()
    plt.show()
```
